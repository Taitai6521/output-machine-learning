```python
X = df_ad.drop(['Admission Chance', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research'],axis=1)
y = df_ad['Admission Chance']
```

```python
X = df_ad[['TOEFL Score','GRE Score']]
y = df_ad['Admission Chance']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(fit_intercept=True)
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
k=X_test.shape[1]
n=len(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_pred)) , '.3f'))
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
MAPE = np.mean( np.abs((y_test - y_pred) /y_test ) ) * 100

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2, '\nMean Absolute Percentage Error =', MAPE, '%') 
```
